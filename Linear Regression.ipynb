## An example: visualizing data in the notebook ✨

Below is an example of a code cell. We'll visualize some simple data using two popular packages in Python. We'll use [NumPy](https://numpy.org/) to create some random data, and [Matplotlib](https://matplotlib.org) to visualize it.

Note how the code and the results of running the code are bundled together.

# This is a simple code where X represents the temperature and Y represents rate of product formation.Linear Regression is done using Gradient descent and 
#Linear Regression model already present.

```python
# Gradient descent method
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from matplotlib import pyplot as plt
X=np.array([40, 45, 50, 55, 60, 65, 70, 75, 80, 85])
Y=np.array([0.15,0.2,0.25,0.27,0.3,0.32,0.35,0.37,0.4,0.42])
plt.scatter(X,Y)
plt.show()
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,shuffle=True,random_state=42)
alpha=0.01 # learning rate
w=0
b=0
epochs=1000
n=len(X_train)
X_train_norm=(X_train-np.mean(X_train))/np.std(X_train)
X_test_norm=(X_test-np.mean(X_test))/np.std(X_test)

for i in range(epochs):
    Y_Pred=w*X_train_norm+b
    error=Y_Pred-Y_train
    dw=(1/n)*np.dot(error,X_train_norm)
    db=(1/n)*np.sum(error)
    w=w-(alpha*dw)
    b=b-(alpha*db)
print(w)
print(b)
Y_test_pred=w*X_test_norm+b
mse=mean_squared_error(Y_test,Y_test_pred)
r2=r2_score(Y_test,Y_test_pred)
print("Mean Squared error: ",mse)
print("R2",r2)
plt.scatter(X_test,Y_test,color='red',label='Actual')
plt.plot(X_test,Y_test_pred,color='green',label='predicted')
```

```python
# Using Linear Regression model 
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
X=np.array([40, 45, 50, 55, 60, 65, 70, 75, 80, 85]).reshape(-1,1)
Y=np.array([0.15,0.2,0.25,0.27,0.3,0.32,0.35,0.37,0.4,0.42])
model=LinearRegression()
model.fit(X,Y)
predicted_Y=model.predict(X)
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)

print("Mean Squared Error (MSE):", mean_squared_error(Y, predicted_Y))
print("R² Score:", r2_score(Y, predicted_Y))

plt.scatter(X,Y,color='red',label='Actual')
plt.plot(X,predicted_Y,color='green',label='predicted')

plt.legend()




```
